```r
library(googledrive)

deliveries_file <- drive_get("NorthStar_dataset/deliveries.csv")

drive_download(
  deliveries_file,
  path = "deliveries.csv",
  overwrite = TRUE
)

deliveries <- read.csv("deliveries.csv")

head(deliveries)
```

```r
library(sqldf)

sqldf("
SELECT delivery_status,
AVG(customer_rating_post_delivery) AS average_rating
FROM deliveries
GROUP BY delivery_status
")
```

The SQL analysis shows that delivery performance has a strong impact on customer satisfaction. Deliveries completed on time received the highest average customer rating of 4.28, while delayed and failed deliveries received much lower ratings of 3.11 and 3.04. This suggests that delays and failed deliveries negatively affect customer experience and reduce the overall service quality within NorthStar’s logistics operations.


```r
sqldf("
SELECT delivery_status,
AVG(manual_route_override_count) AS average_override
FROM deliveries
GROUP BY delivery_status
")
```

The SQL analysis shows that failed and delayed deliveries had higher average manual route override counts compared to deliveries completed on time. This suggests that drivers are more likely to change routes manually during problematic deliveries. The findings may indicate issues such as poor route planning, traffic problems or operational efforts to manage delivery targets. Overall, the results support management concerns about route allocation inefficiencies within NorthStar’s operational network.


```r
sqldf("
SELECT hub_id,
AVG(fuel_or_charge_cost) AS average_cost
FROM deliveries
GROUP BY hub_id
ORDER BY average_cost DESC
")
```

The SQL analysis showed differences in average fuel and charging costs between operational hubs. Hub H05 had the highest average operational cost, while H08 had the lowest. These differences may indicate issues such as inefficient vehicle allocation, higher charging requirements or increased operational expenses at certain hubs. The findings also support management concerns about rising operational costs and inconsistent performance across NorthStar’s logistics network.


```r
hubs_file <- drive_get("NorthStar_dataset/hubs.csv")

drive_download(
  hubs_file,
  path = "hubs.csv",
  overwrite = TRUE
)

hubs <- read.csv("hubs.csv")

head(hubs)
```

```r
sqldf("
SELECT h.zone,
COUNT(*) AS total_delayed

FROM deliveries d

JOIN hubs h
ON d.hub_id = h.hub_id

WHERE d.delivery_status = 'Delayed'

GROUP BY h.zone

ORDER BY total_delayed DESC
")
```

The SQL JOIN analysis shows that the Central zone had the highest number of delayed deliveries, with 47 delayed cases recorded. This supports the earlier Python analysis, where the Central zone also had the highest number of failed deliveries. The repeated results suggest that issues such as route congestion, heavy hub workload or poor operational planning may be more common in the Central area.
